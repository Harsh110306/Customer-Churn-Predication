```python
# Customer Churn Prediction System - Part 1: Exploratory Da...
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os


sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
plt.rcParams["font.size"] = 12
```

```python
# 1. Data Ingestion & Understanding
raw_data_path = "../data/raw/Telco_customer_churn.csv"
df = pd.read_csv(raw_data_path)
print(f"Dataset Shape: {df.shape}")
```

```python
# 1. Data Ingestion & Understanding - Part 5
print("Missing values per column:")
print(df.isnull().sum())
```

```python
# 1. Data Ingestion & Understanding - Part 6
print(f"Duplicate records found: {df.duplicated().sum()}")
```

```python
# 1. Data Ingestion & Understanding - Part 7
df['Total Charges'] = pd.to_numeric(df['Total Charges'].astype(str).str.strip(), errors='coerce')
print(f"Missing values in Total Charges after coercion: {df['Total Charges'].isnull().sum()}")
df['Total Charges'] = df['Total Charges'].fillna(0.0)
```

```python
# Plot 1: Churn Distribution
sns.countplot(data=df, x='Churn Value', palette=['#3b82f6', '#ef4444'])
plt.title('Distribution of Customer Churn')
plt.xlabel('Churn Value (0 = Stayed, 1 = Churned)')
plt.ylabel('Count')
plt.show()
```

```python
# Plot 2: Gender vs Churn
sns.countplot(data=df, x='Gender', hue='Churn Value', palette=['#3b82f6', '#ef4444'])
plt.title('Churn Count by Gender')
plt.show()
```

```python
# Plot 3: Senior Citizen vs Churn
sns.barplot(data=df, x='Senior Citizen', y='Churn Value', palette=['#3b82f6', '#ef4444'], errorbar=None)
plt.title('Churn Rate by Senior Citizen Status')
plt.ylabel('Churn Rate')
plt.show()
```

```python
# Plot 4: Contract Type vs Churn
sns.barplot(data=df, x='Contract', y='Churn Value', palette='Blues_r', errorbar=None)
plt.title('Churn Rate by Contract Type')
plt.ylabel('Churn Rate')
plt.show()
```

```python
# Plot 5: Tenure Months vs Churn
sns.histplot(data=df, x='Tenure Months', hue='Churn Value', multiple='stack', palette=['#3b82f6', '#ef4444'], kde=True)
plt.title('Tenure Months Distribution by Churn Status')
plt.show()
```

```python
# Plot 6: Internet Service vs Churn
sns.barplot(data=df, x='Internet Service', y='Churn Value', palette='Greens_r', errorbar=None)
plt.title('Churn Rate by Internet Service Type')
plt.ylabel('Churn Rate')
plt.show()
```

```python
# Plot 7: Tech Support vs Churn
sns.barplot(data=df, x='Tech Support', y='Churn Value', palette='Oranges_r', errorbar=None)
plt.title('Churn Rate by Tech Support Subscription')
plt.ylabel('Churn Rate')
plt.show()
```

```python
# Plot 8: Payment Method vs Churn
sns.barplot(data=df, x='Payment Method', y='Churn Value', palette='Purples_r', errorbar=None)
plt.xticks(rotation=15)
plt.title('Churn Rate by Payment Method')
plt.ylabel('Churn Rate')
plt.show()
```

```python
# Plot 9: Monthly Charges vs Churn
sns.kdeplot(data=df, x='Monthly Charges', hue='Churn Value', fill=True, palette=['#3b82f6', '#ef4444'])
plt.title('Monthly Charges Density by Churn Status')
plt.show()
```

```python
# Plot 10: Total Charges vs Churn
sns.boxplot(data=df, x='Churn Value', y='Total Charges', palette=['#3b82f6', '#ef4444'])
plt.title('Total Charges Distribution by Churn Status')
plt.show()
```

```python
# Plot 11: CLTV vs Churn
sns.violinplot(data=df, x='Churn Value', y='CLTV', palette=['#3b82f6', '#ef4444'])
plt.title('CLTV Distribution by Churn Status')
plt.show()
```

```python
# Plot 12: Correlation Heatmap
num_cols = ['Tenure Months', 'Monthly Charges', 'Total Charges', 'CLTV', 'Churn Value']
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features')
plt.show()
```
